```python
import pandas as pd
```

```python
df = pd.read_csv("/content/data.csv", encoding='latin-1')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size and style
plt.figure(figsize=(12, 6))

# Calculate the top 10 items based on the total quantity sold
# Use 'df' instead of 'data' to refer to the DataFrame
top_items = df.groupby('Description')['Quantity'].sum().nlargest(10)

# Create a bar plot for the top 10 selling items
sns.barplot(x=top_items.index, y=top_items.values, palette='Blues_d')

# Improve the display: adjust text size and rotate x-axis labels
plt.xticks(size=12, rotation=45, ha='right')
plt.title('Top 10 Best Selling Items', size=16)

# Show the plot
plt.show()
```

```python
# Assuming 'InvoiceDate' column contains date information
# Convert 'InvoiceDate' to datetime objects if it's not already
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract the month from 'InvoiceDate' and create a new 'month' column
df['month'] = df['InvoiceDate'].dt.month

# Now you can proceed with the plotting code:

# Set figure size
plt.figure(figsize=(12, 6))

# Count the number of transactions per month and sort by month order
monthly_transactions = df['month'].value_counts().sort_index()

# Create a bar plot for the number of transactions per month
sns.barplot(x=monthly_transactions.index, y=monthly_transactions.values, palette='Purples_d')

# Improve the display: adjust text size
plt.xticks(size=12)
plt.title('Number of Transactions Each Month', size=16)

# Show the plot
plt.show()
```

```python
# Assuming 'InvoiceDate' column contains date information
# Convert 'InvoiceDate' to datetime objects if it's not already
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract the day of the week from 'InvoiceDate' and create a new 'day_name' column
df['day_name'] = df['InvoiceDate'].dt.day_name()  # Extract day name

# Set figure size
plt.figure(figsize=(12, 6))

# Count the number of transactions for each day of the week
# Use 'df' instead of 'data' to access the DataFrame
daily_transactions = df['day_name'].value_counts()

# Create a bar plot for the number of transactions by day of the week
sns.barplot(x=daily_transactions.index, y=daily_transactions.values, palette='Greens_d')

# Improve the display: adjust text size
plt.xticks(size=12)
plt.title('Number of Transactions by Day of the Week', size=16)

# Show the plot
plt.show()
```

```python
# Assuming 'df' is the DataFrame loaded from the CSV
# Replace 'data' with 'df'
# Check if the column name is 'Member_number' or something similar,
# e.g., 'CustomerID' or 'Customer_ID'. If it's different,
# replace 'Member_number' with the correct column name.
item_count = df.groupby(['CustomerID', 'InvoiceDate', 'Description'])['Description'].count().reset_index(name='Count')

# Display the first 10 rows to verify
print(item_count.head(10))
```

```python
# Generate a 'Transaction' column with unique transaction numbers for each (CustomerID, InvoiceDate) pair
# Replace 'Member_number' and 'Date' with 'CustomerID' and 'InvoiceDate'
item_count['Transaction'] = item_count.groupby(['CustomerID', 'InvoiceDate'])['InvoiceDate'].ngroup() + 1

# Display the first 10 rows to verify
print(item_count.head(10))
```

```python
# Replace 'itemDescription' with 'Description' in the columns parameter
item_count_pivot = item_count.pivot_table(index='Transaction', columns='Description', values='Count', aggfunc='sum').fillna(0)
# drop comma
item_count_pivot = item_count_pivot.astype('int32')
item_count_pivot
```

```python
from mlxtend.frequent_patterns import association_rules, apriori

# Replace 'itemDescription' with 'Description'
basket = item_count.groupby('Transaction')['Description'].apply(list).reset_index()
basket_encoded = basket['Description'].str.join('|').str.get_dummies('|')
basket_encoded
```

```python
frequent_itemsets = apriori(basket_encoded, min_support=0.01, use_colnames=True)
frequent_itemsets.sort_values('support', ascending=False).head(10)

```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=0.5)
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values('confidence', ascending=False).head(10)
```
